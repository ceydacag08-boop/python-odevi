```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

## Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.  

R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  
Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.  
Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar. 
State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  
Profit (Kâr): Şirketin elde ettiği toplam kâr.  
Bu veri seti, startup'ların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz   etmek için kullanılabilir. Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini   incelemek için uygun bir veri setidir.  

## 1.GÖREV : R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
from IPython.display import display

# R&D Harcaması ve Kâr sütunlarını 
rd_kar = data[['R&D Spend', 'Profit']]
display(rd_kar.head(10))

print("\n TEMEL İSTATİSTİKLER ")
display(rd_kar.describe())

print("\n KORELASYON ")
correlation = data['R&D Spend'].corr(data['Profit'])
print(f"R&D Harcaması ve Kâr arasındaki korelasyon: {correlation:.3f}")
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(data['R&D Spend'], data['Profit'], alpha=0.7, color='blue', s=60)
plt.xlabel('R&D Harcaması ($)', fontsize=12)
plt.ylabel('Kâr ($)', fontsize=12)
plt.title('R&D Harcaması ve Kâr ', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 2.GÖREV: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot): Yönetim harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
from IPython.display import display

# Yönetim Harcaması ve Kâr sütunları
print(" YÖNETİM HARCAMASI VE KÂR VERİLERİ ")
yon_kar = data[['Administration', 'Profit']]
display(yon_kar.head(10))

print("\n TEMEL İSTATİSTİKLER ")
display(yon_kar.describe())

print("\n KORELASYON ")
correlation = data['Administration'].corr(data['Profit'])
print(f"Yönetim Harcaması ve Kâr arasındaki korelasyon: {correlation:.3f}")

print("\n EN YÜKSEK VE EN DÜŞÜK DEĞERLER ")
print("En yüksek yönetim harcaması:")
print(data.loc[data['Administration'].idxmax(), ['Administration', 'Profit']])
print("\nEn düşük yönetim harcaması:")
print(data.loc[data['Administration'].idxmin(), ['Administration', 'Profit']])
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(data['Administration'], data['Profit'], alpha=0.7, color='red', s=60)
plt.xlabel('Yönetim Harcaması ($)', fontsize=12)
plt.ylabel('Kâr ($)', fontsize=12)
plt.title('Yönetim Harcamaları ve Kâr', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startup'ların ortalama kârlarını karşılaştıran bir çubuk grafik.

```python

print(" EYALET DAĞILIMI ")
eyalet_dagilim = data['State'].value_counts()
print(eyalet_dagilim)

print("\n EYALETLERE GÖRE KÂR İSTATİSTİKLERİ ")
eyalet_istatistik = data.groupby('State')['Profit'].agg(['count', 'mean', 'median', 'std', 'min', 'max'])
print(eyalet_istatistik)

print("\n EYALETLERE GÖRE ORTALAMA SIRALI KÂR ")
ortalama_kar = data.groupby('State')['Profit'].mean().sort_values(ascending=False)
print(ortalama_kar)

print("\n EN YÜKSEK VE EN DÜŞÜK KÂRLI EYALETLER ")
print(f"En yüksek ortalama kâr: {ortalama_kar.index[0]} - ${ortalama_kar.iloc[0]:,.2f}")
print(f"En düşük ortalama kâr: {ortalama_kar.index[-1]} - ${ortalama_kar.iloc[-1]:,.2f}")
print(f"Fark: ${ortalama_kar.iloc[0] - ortalama_kar.iloc[-1]:,.2f}")
```

```python

plt.figure(figsize=(12, 6))
ortalama_kar = data.groupby('State')['Profit'].mean().sort_values(ascending=False)
bars = plt.bar(ortalama_kar.index, ortalama_kar.values, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])

plt.xlabel('Eyalet', fontsize=12)
plt.ylabel('Ortalama Kâr ($)', fontsize=12)
plt.title('Eyaletlere Göre Ortalama Kâr Karşılaştırması', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

# Değerleri çubukların üzerine yazmak için
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1000,
             f'${height:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

```

## 4. GÖREV: Harcama Türlerinin Karşılaştırması (Boxplot): R&D, yönetim ve pazarlama harcamalarının dağılımını karşılaştıran bir kutu grafiği.

```python

print(" HARCAMA VERİLERİ ")
harcama_verileri = data[['R&D Spend', 'Administration', 'Marketing Spend']]
display(harcama_verileri.head(10))

print("\n HARCAMA İSTATİSTİKLERİ ")
display(harcama_verileri.describe())

print("\n HARCAMA TOPLAMLARI ")
print(f"Toplam R&D harcaması: ${data['R&D Spend'].sum():,.2f}")
print(f"Toplam Yönetim harcaması: ${data['Administration'].sum():,.2f}")
print(f"Toplam Pazarlama harcaması: ${data['Marketing Spend'].sum():,.2f}")

print("\n ORTALAMA DEĞERLER ")
print(f"Ortalama R&D harcaması: ${data['R&D Spend'].mean():,.2f}")
print(f"Ortalama Yönetim harcaması: ${data['Administration'].mean():,.2f}")
print(f"Ortalama Pazarlama harcaması: ${data['Marketing Spend'].mean():,.2f}")

print("\n EN YÜKSEK DEĞERLER ")
print(f"En yüksek R&D harcaması: ${data['R&D Spend'].max():,.2f}")
print(f"En yüksek Yönetim harcaması: ${data['Administration'].max():,.2f}")
print(f"En yüksek Pazarlama harcaması: ${data['Marketing Spend'].max():,.2f}")
```

```python

plt.figure(figsize=(12, 8))

# Veriyi boxplot için hazırlamak 
harcama_data = [data['R&D Spend'], data['Administration'], data['Marketing Spend']]
labels = ['R&D Harcaması', 'Yönetim Harcaması', 'Pazarlama Harcaması']

box_plot = plt.boxplot(harcama_data, tick_labels=labels, patch_artist=True, 
                       boxprops=dict(facecolor='lightblue', alpha=0.7),
                       medianprops=dict(color='red', linewidth=2))


plt.ylabel('Harcama Miktarı ($)', fontsize=12)
plt.title('Harcama Türlerinin Dağılım Karşılaştırması', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')


print("\nHarcama İstatistikleri:")
print(f"R&D Harcaması - Ortalama: ${data['R&D Spend'].mean():,.0f}, Medyan: ${data['R&D Spend'].median():,.0f}")
print(f"Yönetim Harcaması - Ortalama: ${data['Administration'].mean():,.0f}, Medyan: ${data['Administration'].median():,.0f}")
print(f"Pazarlama Harcaması - Ortalama: ${data['Marketing Spend'].mean():,.0f}, Medyan: ${data['Marketing Spend'].median():,.0f}")

plt.tight_layout()
plt.show()
```
